```python
pivot1 = df.pivot_table(values='Survived', index='Sex', columns='Pclass', aggfunc='mean')
print(pivot1)
```

```python
pivot2 = df.pivot_table(values='Age', index='Sex', columns='Survived', aggfunc='mean')
print(pivot2)
```

```python
pivot3 = df.pivot_table(values='Survived', index='Sex', columns='Pclass', aggfunc='mean', margins=True)
print(pivot3)
```

```python
pivot_sibsp = df.pivot_table(values='Survived', index='SibSp', aggfunc='mean')
print(pivot_sibsp)
```

```python
pivot5 = df.pivot_table(values='PassengerId', index='Embarked', columns='Pclass', aggfunc='count')
print(pivot5)
```

```python
pivot_parch = df.pivot_table(values='Survived', index='Parch', aggfunc='mean')
print(pivot_parch)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())  
df
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=range(0, 85, 5), right=False)

age_survival = df.groupby('AgeGroup').agg(
    SurvivalRate=('Survived', 'mean'),
    Count=('Survived', 'size')
)

age_survival

```

```python
df_dummies = pd.get_dummies(df, columns=['Sex', 'Embarked', 'Pclass'], drop_first=True)


print(df_dummies.head())
```
