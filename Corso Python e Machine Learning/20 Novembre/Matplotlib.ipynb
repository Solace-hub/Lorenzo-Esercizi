```python
x = np.linspace(0, 2 * np.pi, 10)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.title("Sine Wave")
plt.xlabel("x values from 0 to 2pi")
plt.ylabel("sin(x)")
plt.show
```

```python
plt.scatter(x,y)
```

```python
categories = ["A", "B", "C", "D"]
values = [23,17,35,29]

plt.bar(categories, values, color = ["red", "blue", "green", "purple"])
plt.title("Bar chart example")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Programming Language Popularity")
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, y, label='Sine Wave')
plt.title("Annotated Sine Wave")
plt.xlabel("x values")
plt.ylabel("sin(x)")
plt.legend()

peak_x = np.pi / 2
peak_y = 1
plt.annotate('Peak', xy=(peak_x, peak_y), xytext=(peak_x + 1, peak_y - 0.5),
arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('x values')
ax1.set_ylabel('sin(x)', color=color)
ax1.plot(x, y, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('cos(x)', color=color)
ax2.plot(x, np.cos(x), color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Dual Axes Example")
fig.tight_layout() # To prevent labels from overlapping
plt.show()
```
