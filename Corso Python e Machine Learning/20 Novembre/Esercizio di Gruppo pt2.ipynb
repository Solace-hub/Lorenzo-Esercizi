```python
df.dropna(subset=['fuel_type', 'clean_title', 'accident'], inplace=True)
df.isnull().sum()
```

```python
threshold = 0.6 * len(df)
cols_with_nulls = df.columns[df.isnull().mean() > 0.6]
df.drop(columns=cols_with_nulls, inplace=True)
```

```python
df_pulito = df[df['int_col'] != '-']
df_pulito
```

```python
df_pulito.to_csv('dataframe_pulito.csv', index=False)
```

```python
df_pulito = pd.read_csv("dataframe_pulito.csv")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_pulito['milage'], df_pulito['price'], alpha=0.5, c='green', edgecolor='black')
plt.title("Prezzo vs Chilometraggio")
plt.xlabel("Chilometraggio")
plt.ylabel("Prezzo")
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
boxplot_data = [df_pulito[df_pulito['model_year'] == year]['price'] for year in sorted(df_pulito['model_year'].dropna().unique())]
plt.boxplot(boxplot_data, showfliers=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Prezzo vs Anno di Produzione")
plt.xlabel("Anno di Produzione")
plt.ylabel("Prezzo")
plt.xticks(ticks=range(1, len(boxplot_data) + 1), labels=sorted(df_pulito['model_year'].dropna().unique()), rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
