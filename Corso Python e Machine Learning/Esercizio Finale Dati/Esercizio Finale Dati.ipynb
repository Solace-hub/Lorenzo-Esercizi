```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Generazione dei Dati: utilizzare Numpy per generare una serie temporale di 305 giorni di dati, simulando il numero di visitatori giornalieri in un ospedale. 
# Assumere che il numero medio di visitatori sia 1200 con una deviazione standard di 900. 
# Inoltre, aggiungere un trend decrescente nel tempo per simulare l'aumento della popolarità del parco.


giorni = 365
media_visite = 1200
deviazione_standard = 900
trend_decrescente = 0.5
```

```python
array_giorni = np.arange(giorni)
array_giorni
```

```python
visite_giornaliere = np.random.normal(loc = media_visite, scale = deviazione_standard, size = giorni)
visite_giornaliere
```

```python
visite_giornaliere -= array_giorni * trend_decrescente
visite_giornaliere
```

```python
# Creazione del dataframe: Creare un Dataframe con Pandas in cui le date devono fare da indice e il numero dei visitatori devono essere le colonna.
# In più, deve esserci una colonna casuale per una patologia del cuore. (Qualsiasi patologia)

patologie_cardiache = ["Infarto", "Ictus", "Insufficienza cardiaca"]
```

```python
patologia_casuale = np.random.choice(patologie_cardiache, size = giorni)
```

```python
date = pd.date_range(start = "25-09-2024", periods = giorni, freq = "D") #D sta per daily, giornaliero
```

```python
df = pd.DataFrame({
    "Visite" : visite_giornaliere,
    "Patologia" : patologia_casuale
}, index = date)

df
```

```python
# Analisi dei Dati: Calcolare il numero medio di visitatori per mese e la
# deviazione standard quindi quale patologia è quella più trovata e quale quella di meno

df["Mese"] = df.index.month
visite_mese = df.groupby("Mese")["Visite"].agg(["mean", "std" ])
visite_mese
```

```python
patologia_frequente = df["Patologia"].mode()[0] #Mode restituisce il valore più frequente
patologia_occasionale = df["Patologia"].value_counts().idxmin() #Indice minimo di occorrenze 
```

```python
print(patologia_frequente)
print(patologia_occasionale)
```

```python
# Visualizzazione dei dati: creazione del grafico a linee del numero di visitatori giornalieri

df["Media_Mobile_7Giorni"] = df["Visite"].rolling(window=7, min_periods=1).mean() #Rolling, specifica una finestra di 7 giorni

plt.figure(figsize=(10, 6))

# Grafico dei visitatori giornalieri
plt.plot(df.index, df["Visite"], label="Visitatori Giornalieri", color="blue")

# Grafico della media mobile a 7 giorni
plt.plot(df.index, df["Media_Mobile_7Giorni"], label="Media Mobile 7 Giorni", color="orange", linestyle="--")
plt.title("Numero di Visitatori Giornalieri con Media Mobile a 7 Giorni")
plt.xlabel("Data")
plt.ylabel("Numero di Visitatori")
plt.xticks(rotation=45)


plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Creare un secondo grafico che mostri la media mensile dei visitatori.

df["Mese"] = df.index.month

media_mensile = df.groupby("Mese")["Visite"].mean()

plt.figure(figsize=(10, 6))

plt.plot(media_mensile.index, media_mensile, label="Media Mensile Visitatori", color="green", marker="o")

plt.title("Media Mensile dei Visitatori")
plt.xlabel("Mese")
plt.ylabel("Numero di Visitatori")
#plt.xticks(rotation=45)

plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Creare un grafico che mostri la divisione fra le 3 patologie.

patologie_count = df["Patologia"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(patologie_count, labels=patologie_count.index, autopct="%1.1f%%", startangle=90, colors=["red", "blue", "lime"]) 
plt.title("Distribuzione delle Patologie Cardiache")
plt.show()
```
