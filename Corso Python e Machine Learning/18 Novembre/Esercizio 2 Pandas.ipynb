```python
import numpy as np
import pandas as pd
```

```python
prodotti = ['Prodotto A', 'Prodotto B', 'Prodotto C', 'Prodotto D']
città = ['Roma', 'Milano', 'Napoli', 'Torino']
```

```python
dati = {
    'Prodotto': np.random.choice(prodotti, 20),
    'Quantità': np.random.randint(1, 20, 20),
    'Prezzo Unitario': np.random.uniform(10, 100, 20).round(2),
    'Città': np.random.choice(città, 20)
}

df = pd.DataFrame(dati)
```

```python
df['Totale Vendite'] = df['Quantità'] * df['Prezzo Unitario']
df
```

```python
totale_vendite_per_prodotto = df.groupby('Prodotto')['Totale Vendite'].sum()
totale_vendite_per_prodotto
```

```python
prodotto_più_venduto = df.groupby('Prodotto')['Quantità'].sum().idxmax()
prodotto_più_venduto
```

```python
città_con_maggiore_vendite = df.groupby('Città')['Totale Vendite'].sum().idxmax()
città_con_maggiore_vendite
```

```python
vendite_superiori_1000 = df[df['Totale Vendite'] > 1000]
vendite_superiori_1000
```

```python
df_ordinato = df.sort_values(by='Totale Vendite', ascending=False)
df_ordinato
```

```python
numero_vendite_per_città = df['Città'].value_counts()
numero_vendite_per_città
```
